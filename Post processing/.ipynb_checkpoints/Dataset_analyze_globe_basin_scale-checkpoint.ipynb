```python
# Extract unique basin names and IDs
Basin_names=china_basins['MAJ_NAME'].to_list()
Basin_IDs=china_basins['MAJ_BAS'].to_list()
Basin_name_unique=[]
Basin_ID_unique=[]
Basin_ID_unique=list(set(Basin_IDs))
Basin_name_unique=list(set(Basin_names))
```

```python
# This code performs parallel processing to analyze basin-level
# water level fluctuations and change rates, 
# by merging VS stages to construct basin-level time series.

import multiprocessing as mp
from tqdm.notebook import tqdm
import warnings
from mpfun import process_single_basin, init_worker
warnings.filterwarnings('ignore')

# Define global variables to be shared among all processes
global_vars = None

# Prepare shared data for parallel processing
shared_data = {
    'Basin_IDs': Basin_IDs,
    'china_basins': china_basins,
    'lons': lons,
    'lats': lats,
    'fluctuation_np': fluctuation_np,
    'change_rate_np': change_rate_np,
    'new_dataset2': new_dataset2
}

# Set the number of processes to be used
n_processes = max(1, 5)  # Leave one core for the system

# Initialize result lists
Basin_collect_name = []
Basin_collect_fluctuation = []
Basin_collect_change_rate = []
Basin_collect_fluctuation2 = []
Basin_collect_change_rate2 = []
Basin_collect_fluctuation_median = []
Basin_collect_change_rate_median = []

# Perform parallel processing using Pool
try:
    with mp.Pool(processes=n_processes, initializer=init_worker, initargs=(shared_data,)) as pool:
        results = list(tqdm(
            pool.imap(process_single_basin, Basin_ID_unique),
            total=len(Basin_ID_unique),
            desc="Processing basins"
        ))
except Exception as e:
    print(f"Error in parallel processing: {str(e)}")
    results = []

# Process the results
for result in results:
    if result is not None:
        Basin_collect_name.append(result['basin_name'])
        Basin_collect_fluctuation.append(result['fluctuation'])
        Basin_collect_change_rate.append(result['change_rate'])
        Basin_collect_fluctuation2.append(result['fluctuation2'])
        Basin_collect_change_rate2.append(result['change_rate2'])  # m/yr
        Basin_collect_fluctuation_median.append(result['fluctuation_median'])
        Basin_collect_change_rate_median.append(result['change_rate_median'])

print(f"Processing complete! Obtained data for {len(Basin_collect_name)} basins.")
```

```python
# Create an empty GeoDataFrame with additional columns for calculated basin-level results
# This variable stores calculated results for global basins
collected_basins = gpd.GeoDataFrame(columns=china_basins.columns.tolist() + ['median_fluctuation','fluctuation','median_change_rate','change_rate','fluctuation2','change_rate2'])
#fluctuation_sorted_indices=np.argsort(Basin_collect_fluctuation)

for i in range(len(Basin_collect_name)):
    selected_basins=china_basins[china_basins['MAJ_NAME']==Basin_collect_name[i]]
    # If selected basins contain multiple boundaries (polygons), their geometries were combined into one unified geometry
    combined_geometry = unary_union(selected_basins.geometry)
    # Copy the first record of the selected basins and update its geometry
    new_record = selected_basins.iloc[0].copy()
    new_record.geometry = combined_geometry
    new_record['median_fluctuation'] = Basin_collect_fluctuation_median[i]
    new_record['fluctuation'] = Basin_collect_fluctuation[i]
    new_record['median_change_rate'] = Basin_collect_change_rate_median[i]
    new_record['change_rate'] = Basin_collect_change_rate[i]
    new_record['fluctuation2'] = Basin_collect_fluctuation2[i]
    new_record['change_rate2'] = Basin_collect_change_rate2[i]
    # Append the new record to the collected basins GeoDataFrame
    collected_basins = pd.concat([collected_basins, gpd.GeoDataFrame([new_record])], ignore_index=True)

# Convert the unit
collected_basins['median_change_rate']=collected_basins['median_change_rate']*100#->cm/year
collected_basins['change_rate2']=collected_basins['change_rate2']*100#->cm/year
```

## Analyze water level seasonal variations by basins

```python
# Parallel processing of seasonal analysis at the basin scale
# - Sets up multiprocessing environment for efficient basin processing
# - Calculates seasonal statistics for each basin

import multiprocessing as mp
from tqdm.notebook import tqdm
import warnings
from mpfun_season import process_single_basin,init_worker
warnings.filterwarnings('ignore')

# Initialize global variables for sharing across processes
global_vars = None

# Prepare shared data for parallel processing
shared_data = {
    'Basin_IDs': Basin_IDs,
    'china_basins': china_basins,
    'lons': lons,
    'lats': lats,
    'new_dataset2': new_dataset2
}

# Configure parallel processing
n_processes = max(1, 5)  # Reserve one core for system

# Initialize result collection lists
Basin_collect_name2=[]#Basin name
Basin_collect_AE_median2=[]#Median AE 
Basin_collect_AE_slope_median2=[]#Basin-level AE change rate
Basin_collect_year_max_slope_median2=[]#Basin-level change rate of seasonal maximum stages
Basin_collect_year_min_slope_median2=[]#Basin-level change rate of seasonal minimum stages
Basin_collect_fluctuation_slope_median2=[]#Basin-level change rate of seasonal fluctuations

# Execute parallel processing
try:
    with mp.Pool(processes=n_processes, initializer=init_worker, initargs=(shared_data,)) as pool:
        results = list(tqdm(
            pool.imap(process_single_basin, Basin_ID_unique),
            total=len(Basin_ID_unique),
            desc="Processing basins"
        ))
except Exception as e:
    print(f"Error in parallel processing: {str(e)}")
    results = []

# Collect and organize results
for result in results:
    if result is not None:
        Basin_collect_name2.append(result['basin_name'])
        
        Basin_collect_AE_median2.append(result['AE_median'])
        Basin_collect_AE_slope_median2.append(result['AE_slope'])
        Basin_collect_year_max_slope_median2.append(result['year_max_slope'])
        Basin_collect_year_min_slope_median2.append(result['year_min_slope'])
        Basin_collect_fluctuation_slope_median2.append(result['year_fluctuation_slope'])

print(f"Processing complete! Analyzed {len(Basin_collect_name2)} basins.")
```

```python
# Create GeoDataFrame for basin-level seasonal statistics
# - Combines basin geometries with calculated seasonal metrics

rcParams['font.family'] = 'Arial'
collected_basins = gpd.GeoDataFrame(columns=china_basins.columns.tolist() + ['median_AE',
                                                                             'median_AE_slope',
                                                                            'median_year_max_slope',
                                                                            'median_year_min_slope',
                                                                            'median_fluctuation_slope'])

for i in range(len(Basin_collect_name2)):
    selected_basins=china_basins[china_basins['MAJ_NAME']==Basin_collect_name2[i]]
    # Combine geometries if basin has multiple polygons
    combined_geometry = unary_union(selected_basins.geometry)
    # Create new record with basin geometry and statistics
    new_record = selected_basins.iloc[0].copy()
    new_record.geometry = combined_geometry
    new_record['median_AE'] = Basin_collect_AE_median2[i]
    new_record['median_AE_slope'] = Basin_collect_AE_slope_median2[i]
    new_record['median_year_max_slope'] = Basin_collect_year_max_slope_median2[i]
    new_record['median_year_min_slope'] = Basin_collect_year_min_slope_median2[i]

    new_record['median_fluctuation_slope'] = Basin_collect_fluctuation_slope_median2[i]
    # Add record to collection
    collected_basins = pd.concat([collected_basins, gpd.GeoDataFrame([new_record])], ignore_index=True)
```

```python
# Convert change rates from m/year to cm/year
collected_basins['median_year_max_slope']=collected_basins['median_year_max_slope']*100
collected_basins['median_year_min_slope']=collected_basins['median_year_min_slope']*100
collected_basins['median_fluctuation_slope']=collected_basins['median_fluctuation_slope']*100
```

```python
"""
Use multiprocessing to analyze extreme years for multiple basins
This script sets up parallel processing to efficiently analyze water level data 
across multiple basins, calculating the year with most extreme stages and that showing
the highest frequency of extreme stages during wet seasons.
"""
import multiprocessing as mp
from tqdm.notebook import tqdm
import warnings
from mpfun_extreme_year2 import process_single_basin,init_worker
warnings.filterwarnings('ignore')

# Define global variable that will be shared across all processes
global_vars = None

# Prepare shared data dictionary for all processes
shared_data = {
    'Basin_IDs': Basin_IDs,         # Basin identification data
    'china_basins': china_basins,   # Global basin boundaries
    'new_dataset2': new_dataset2,
    'wet_season': {                # Wet season parameters
        'lat': wet_season['lat'].values,
        'lon': wet_season_lon, 
        'start_month': wet_season['start_month'].values,
        'end_month': wet_season['end_month'].values,
        'median_month': wet_season['median_month'].values
    }
}

# Set number of processes
n_processes = max(1, 6)  # Reserve one core for system operations

# Initialize result lists for collecting basin statistics
Basin_collect_name2=[] #Basin names
Basin_collect_low_year_percent_most_whole_year=[]# Low stage year (whole year)
Basin_collect_high_year_percent_most_whole_year=[]# High stage year (whole year)

Basin_collect_low_year_percent_most_wet_season=[]# Low stage year (wet season)
Basin_collect_high_year_percent_most_wet_season=[]# High stage year (wet season)

# Execute parallel processing using Pool
try:
    with mp.Pool(processes=n_processes, initializer=init_worker, initargs=(shared_data,)) as pool:
        results = list(tqdm(
            pool.imap(process_single_basin, Basin_ID_unique),
            total=len(Basin_ID_unique),
            desc="Processing basins"
        ))
except Exception as e:
    print(f"Error in parallel processing: {str(e)}")
    results = []

# Process and collect results from all completed basin analyses
for result in results:
    if result is not None:
        Basin_collect_name2.append(result['basin_name'])
        
        Basin_collect_low_year_percent_most_whole_year.append(result['low_year_basin_whole_year'])
        Basin_collect_high_year_percent_most_whole_year.append(result['high_year_basin_whole_year'])
        
        Basin_collect_low_year_percent_most_wet_season.append(result['low_year_basin_wet_season'])
        Basin_collect_high_year_percent_most_wet_season.append(result['high_year_basin_wet_season'])
        
print(f"Processing completed! Total number of processed basins: {len(Basin_collect_name)}")
```

```python
#Calculate Year Difference Between Low Stage Year and High stage year

Diff_low_high=[Basin_collect_low_year_percent_most[i]-Basin_collect_high_year_percent_most[i] 
               for i in range(len(Basin_collect_low_year_percent_most))]
len(Diff_low_high)
```

```python
"""
Spatial Visualization of Time Lag Between Extreme Water Level Years
This script creates a map visualization showing the temporal difference (lag) between 
low and high water stage years for global basins. It includes:
1. A main map showing basins colored by year lag
2. An inset histogram showing the distribution of year lags
Positive lag means low stage year occurred after high stage year
Negative lag means high stage year occurred after low stage year
"""
collected_basins = gpd.GeoDataFrame(columns=china_basins.columns.tolist() + ['Basin_low_high_diff'])
# Process each basin: combine geometries and add year lag data
for i in range(len(Basin_collect_name)):
    selected_basins=china_basins[china_basins['MAJ_NAME']==Basin_collect_name[i]]
    combined_geometry = unary_union(selected_basins.geometry)
    
    new_record = selected_basins.iloc[0].copy()
    new_record.geometry = combined_geometry
    new_record['Basin_low_high_diff'] = Diff_low_high[i]
    collected_basins = pd.concat([collected_basins, gpd.GeoDataFrame([new_record])], ignore_index=True)
# Convert year lag column to integer type
collected_basins['Basin_low_high_diff'] = collected_basins['Basin_low_high_diff'].astype(int)

fig, ax2 = plt.subplots(figsize=(12, 5), subplot_kw={'projection': ccrs.PlateCarree()})

cmap = 'RdBu_r'
norm = mcolors.Normalize(vmin=-7, vmax=7)

ax2.coastlines()
ax2.set_global()
ax2.set_extent([lon_min,lon_max,lat_min,lat_max],crs=ccrs.PlateCarree())
collected_basins.plot(ax=ax2, column='Basin_low_high_diff', cmap=cmap, norm=norm, legend=False, edgecolor='k',linewidth=0.3)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm._A = [] 
cbar2 = plt.colorbar(sm, ax=ax2, orientation='horizontal', fraction=0.046, pad=0.04)
cbar2.set_label('Year lag',fontsize=14)
cbar2.ax.tick_params(labelsize=12)
cbar2.ax.xaxis.set_major_locator(MultipleLocator(2))
# Plot histogram of year lags
bins=np.arange(-7.5,7.5,1)
rect9=[0.17,0.37,0.13,0.15]
ax9=fig.add_axes(rect=rect9)
ax9.hist(Diff_low_high,bins=bins,edgecolor='none',color='gray',alpha=0.5)
ax9.set_ylabel('Count', fontname='Arial',fontsize=12,color='k')
ax9.set_xlabel('Year lag', fontname='Arial',fontsize=12,color='k')
ax9.tick_params(labelsize=10,colors='k')
ax9.spines['right'].set_visible(False)
ax9.spines['top'].set_visible(False)
ax9.set_xlim([-7.5,7.5])
ax9.set_facecolor('white')
ax9.text(-7,20,f'>0: {sum(np.array(Diff_low_high)>0)}\n<0: {sum(np.array(Diff_low_high)<0)}'
        ,fontname='Arial', color='k', fontsize=12)

plt.tight_layout()
plt.show()
```

```python
"""
Spatial Visualization of Time Lag Between Extreme Water Level Years
This script creates a map visualization showing the temporal difference (lag) between 
low and high water stage years for global basins. It includes:
1. A main map showing basins colored by year lag
2. An inset histogram showing the distribution of year lags
Positive lag means low stage year occurred after high stage year
Negative lag means high stage year occurred after low stage year
"""
collected_basins = gpd.GeoDataFrame(columns=china_basins.columns.tolist() + ['Basin_low_high_diff'])

for i in range(len(Basin_collect_name)):
    selected_basins=china_basins[china_basins['MAJ_NAME']==Basin_collect_name[i]]
    combined_geometry = unary_union(selected_basins.geometry)
    
    new_record = selected_basins.iloc[0].copy()
    new_record.geometry = combined_geometry
    new_record['Basin_low_high_diff'] = Diff_low_high[i]
    collected_basins = pd.concat([collected_basins, gpd.GeoDataFrame([new_record])], ignore_index=True)

collected_basins['Basin_low_high_diff'] = collected_basins['Basin_low_high_diff'].astype(int)

fig, ax2 = plt.subplots(figsize=(12, 5), subplot_kw={'projection': ccrs.PlateCarree()})

cmap = 'RdBu_r'
norm = mcolors.Normalize(vmin=-7, vmax=7)

ax2.coastlines()
ax2.set_global()
ax2.set_extent([lon_min,lon_max,lat_min,lat_max],crs=ccrs.PlateCarree())
collected_basins.plot(ax=ax2, column='Basin_low_high_diff', cmap=cmap, norm=norm, legend=False, edgecolor='k',linewidth=0.3)
sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm._A = [] 
cbar2 = plt.colorbar(sm, ax=ax2, orientation='horizontal', fraction=0.046, pad=0.04)
cbar2.set_label('Year lag',fontsize=14)
cbar2.ax.tick_params(labelsize=12)
cbar2.ax.xaxis.set_major_locator(MultipleLocator(2))

bins=np.arange(-7.5,7.5,1)
rect9=[0.17,0.37,0.13,0.15]
ax9=fig.add_axes(rect=rect9)
ax9.hist(Diff_low_high,bins=bins,edgecolor='none',color='gray',alpha=0.5)
ax9.set_ylabel('Count', fontname='Arial',fontsize=12,color='k')
ax9.set_xlabel('Year lag', fontname='Arial',fontsize=12,color='k')
ax9.tick_params(labelsize=10,colors='k')
ax9.spines['right'].set_visible(False)
ax9.spines['top'].set_visible(False)
ax9.set_xlim([-7.5,7.5])
ax9.set_facecolor('white')
ax9.text(-5,20,f'>0: {sum(np.array(Diff_low_high)>0)}\n<0: {sum(np.array(Diff_low_high)<0)}'
        ,fontname='Arial', color='k', fontsize=12)

plt.tight_layout()
plt.show()
```
